```python
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])

print("Add:\n", x + y)
print("Sub:\n", x - y)
mul_x = x * 3
print("Multiply x by 3:\n", mul_x)

square_x = np.power(x,2)
print("square\n",square_x)

v = np.array([1, 2, 3])
w = np.array([4, 5, 6])
x = np.array([7, 8, 9])

# v.w
dot_vw = np.dot(v, w)
print("Dot product v路w:", dot_vw)

# x.v
dot_xv = np.dot(x, v)
print("Dot product x路v:", dot_xv)


x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])

xy_row = np.concatenate((x, y), axis=0)
print("Concate:",xy_row)
```

```python
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]]);

A_inv = np.linalg.inv(A)
print("A inverse:\n", A_inv)

Ide = np.dot(A, A_inv)
print("A * A_inv =\n", Ide)


AB = np.dot(A,B)
print("AB doesnot equal to BA\n", AB)

BA = np.dot(B,A)
print("BA doesnot equal to AB\n", BA)


LHS = np.transpose(AB)
print("LHS\n", LHS)

RHS = np.dot(np.transpose(B), np.transpose(A))
print("RHS\n", RHS)
```
