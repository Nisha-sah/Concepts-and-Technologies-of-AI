```python
import pandas as pd
df= pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/medical_students_dataset.csv")
df.head()

```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Columns in dataset:", df.columns)
print("\nMissing values before imputation:\n", df.isnull().sum())

for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ['float64', 'int64']:  # Numerical column
            # Fill numerical missing values with median
            df[col] = df[col].fillna(df[col].median())
            print(f"Filled missing numerical values in '{col}' with median.")
        else:  # Categorical column
            # Fill categorical missing values with mode
            df[col] = df[col].fillna(df[col].mode()[0])
            print(f"Filled missing categorical values in '{col}' with mode.")

print("\nMissing values after imputation:\n", df.isnull().sum())
```

```python
duplicates_before = df.duplicated().sum()
print(f"\nNumber of duplicate rows before removal: {duplicates_before}")

df = df.drop_duplicates()

duplicates_after = df.duplicated().sum()
print(f"Number of duplicate rows after removal: {duplicates_after}")

# Preview cleaned data
print("\nPreview of cleaned dataset:")
print(df.head())
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/Titanic-Dataset.csv")
df.head()
```

```python
import pandas as pd
subset = df[['Name','Pclass','Sex','Age','Fare','Survived']]
subset_1st = subset[subset['Pclass'] == 1]

# calculate statistics
mean_fare = subset_1st['Fare'].mean()
median_fare = subset_1st['Fare'].median()
max_fare = subset_1st['Fare'].max()
min_fare = subset_1st['Fare'].min()



print("Mean Fare:", mean_fare)
print("Median Fare:", median_fare)
print("Max Fare:", max_fare)
print("Min Fare:", min_fare)

```

```python
import pandas as pd

# Count null  Age column
null_count = subset_1st['Age'].isnull().sum()
print("Number of null values in Age:", null_count)

# Drop rows missing Age
subset_age_clean = subset_1st.dropna(subset=['Age'])
print("DataFrame after dropping null Age values:")
print(subset_age_clean)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/drive/MyDrive/Week - 2 - Describing the World with Data/Dataset/Titanic-Dataset.csv")
survival_by_gender = df.groupby('Sex')['Survived'].mean()
print(survival_by_gender)

plt.figure(figsize=(6,4))
sns.barplot(data=df, x='Sex', y='Survived')
plt.title("Mean Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Mean Survival Rate")
plt.show()


```
