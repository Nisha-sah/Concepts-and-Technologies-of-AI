```python
import pandas as pd
total_fare = df['Fare'].sum()
print (total_fare)
```

```python

# First class total fare
fare_first = df[df["Pclass"] == 1]["Fare"].sum()
print(fare_first)

# Second class total fare
fare_second = df[df["Pclass"] == 2]["Fare"].sum()
print(fare_second)

# Third class total fare
fare_third = df[df["Pclass"] == 3]["Fare"].sum()
print(fare_third)

# Combine into a list
fare_list = [fare_first, fare_second, fare_third]
print(fare_list)

```

```python

fare_revenue = [f / total_fare for f in fare_list]
print(fare_revenue)
```

```python

# Step 1
df["Age"].fillna(df["Age"].median(), inplace=True)

df["age_group"] = pd.cut(
    df["Age"],
    bins=[0, 18, 65, 120],
    labels=["child", "adult", "senior"],
    right=False
)

# Step 2
total_passengers = len(df)

# Step 3
age_group_counts = df["age_group"].value_counts()

# Step 4
age_group_proportion = age_group_counts / total_passengers

# Step 5
age_group_percentage = (age_group_proportion * 100).round(2)
age_group_percentage
```
